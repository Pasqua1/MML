##### 12. Как факт выживания зависит от возраста пассажира? Проверьте (графически) предположение, что молодые чаще выживали. Пусть, условно, молодые - младше 30 лет, пожилые – старше 60 лет.

Напишите код и визуализируйте

```python
#ваш код здесь
d=data[['Survived','Age']]
f=d['Age']<=30
young=d[f]
f=d['Age']>60
old=d[f]
k=pd.concat([young,old], ignore_index=True)
f=k['Age']>60
k.insert(2,'flag',list(f),True)
#ax = sns.countplot(x="Age", hue="Survived", data=young)
bx = sns.catplot(x="Age", hue="Survived", col='flag', kind='count', data=k)
```
